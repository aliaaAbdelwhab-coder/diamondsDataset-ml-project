```python
import numpy as np 
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data_Frame = pd.read_csv('..\\diamonds.csv')
data_Frame
```

```python
data_Frame
```

```python
features = data_Frame.drop(['price','Unnamed: 0'], axis=1)
```

```python
features.head()
```

```python
features = pd.get_dummies(features, columns=['cut', 'color', 'clarity'])
```

```python
features.head()
```

```python
features = features.to_numpy()
y = data_Frame['price'].to_numpy()
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(features)
```

```python
[xtrain,xtest,ytrain,ytest]=train_test_split(x_scaled,y,test_size=0.2 )
```

```python
knn = KNeighborsRegressor(n_neighbors=4)
```

```python
pd.DataFrame(xtrain).isnull().sum().sum()

```

```python
model = knn.fit(xtrain, ytrain)
```

```python
predictions = model.predict(xtest)
score = model.score(xtest, ytest)
score
```
